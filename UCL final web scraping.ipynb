```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_Cup_and_UEFA_Champions_League_finals'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
soup
```

```python
table = soup.find('table' , class_="wikitable plainrowheaders sortable")
```

```python
table
```

```python
col_table = table.find('tbody')
```

```python
titles = col_table.find_all('th' , scope = 'col')
```

```python
head = [title.text.strip() for title in titles]
```

```python
head
```

```python
head = head[0:8]
```

```python
head
```

```python
head[-1] = 'Attendance'
```

```python
head
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = head)
df
```

```python
col_data = table.find_all('tr')
```

```python
col_data
```

```python
for row in col_data[1:-4]:
    row_data = row.find_all('td')
    yd = row.find_all('th')
    year_data = [y.text.strip() for y in yd]
    ind_data =year_data+ [data.text.strip()for data in row_data]
    #print(ind_data)
    if len(ind_data) !=8:
        continue
    length = len(df)
    df.loc[length] = (ind_data)
```

```python
df
```

```python
season_to_find = '1973–74'
result = df.loc[df['Season'] == season_to_find]
```
